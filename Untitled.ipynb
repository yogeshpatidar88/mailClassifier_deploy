```python
data.rename(columns={'Text':'text', 'Class':'spam'},inplace=True)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
import string

nltk.download('punkt')
nltk.download('stopwords')

def transform_text(text):
    # Step 1: Convert to lowercase
    text = text.lower()
    
    # Step 2: Tokenize
    text = word_tokenize(text)
    
    # Step 3: Remove non-alphanumeric characters
    text = [word for word in text if word.isalnum()]
    
    # Step 4: Remove stopwords and punctuation
    stop_words = set(stopwords.words('english'))
    text = [word for word in text if word not in stop_words and word not in string.punctuation]
    
    # Step 5: Stemming
    ps = PorterStemmer()
    text = [ps.stem(word) for word in text]
    
    # Step 6: Join tokens back into a single string
    return " ".join(text)
```

```python
data['txt']=data['text'].apply(transform_text)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2 , random_state= 0)
```
